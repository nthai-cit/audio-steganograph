```python


FILE_PATH = "/content/drive/MyDrive/Git-audio-stego/audio-steganograph/Steganalysis/Data/LSB_CNN/features_cnn.npz"

def fix_npz_keys(path):
    print(f"--- ДђANG KIб»‚M TRA FILE: {path} ---")

    if not os.path.exists(path):
        print(" Lб»—i: KhГґng tГ¬m thбєҐy file.")
        return

    try:
        # Load file б»џ chбєї Д‘б»™ Д‘б»Ќc
        with np.load(path) as data:
            keys = list(data.files)
            print(f" CГЎc key hiб»‡n tбєЎi trong file: {keys}")

            # Nбєїu Д‘ГЈ Д‘Гєng chuбє©n thГ¬ dб»«ng
            if 'X' in keys and 'y' in keys:
                print(" File Д‘ГЈ Д‘Гєng chuбє©n (cГі X vГ  y). KhГґng cбє§n sб»­a.")
                return

            # Logic Д‘oГЎn vГ  sб»­a key
            new_data = {}
            found_X = False
            found_y = False

            for k in keys:
                arr = data[k]
                print(f"   -> Key '{k}': shape={arr.shape}, type={arr.dtype}")

                # ДђoГЎn X: ThЖ°б»ќng lГ  mбєЈng nhiб»Ѓu chiб»Ѓu (бєўnh: 4 chiб»Ѓu, Feature: 2 chiб»Ѓu)
                if len(arr.shape) > 1 and not found_X:
                    new_data['X'] = arr
                    found_X = True
                    print(f"      => ДђГЈ map '{k}' thГ nh 'X'")

                # ДђoГЎn y: ThЖ°б»ќng lГ  mбєЈng 1 chiб»Ѓu (Labels)
                elif len(arr.shape) == 1 and not found_y:
                    new_data['y'] = arr
                    found_y = True
                    print(f"      => ДђГЈ map '{k}' thГ nh 'y'")

            # LЖ°u lбєЎi file mб»›i nбєїu tГ¬m Д‘б»§ X vГ  y
            if found_X and found_y:
                print(" Дђang ghi Д‘ГЁ file vб»›i key chuбє©n 'X' vГ  'y'...")
                np.savez_compressed(path, X=new_data['X'], y=new_data['y'])
                print(" ДђГѓ Sб»¬A THГЂNH CГ”NG! BбєЎn cГі thб»ѓ chбєЎy lб»‡nh Train ngay.")
            else:
                print(" KhГґng thб»ѓ tб»± Д‘б»™ng xГЎc Д‘б»‹nh X vГ  y. Vui lГІng xГіa file nГ y vГ  tбєЎo lбєЎi cache tб»« Д‘бє§u.")

    except Exception as e:
        print(f" Lб»—i khi Д‘б»Ќc file: {e}")

if __name__ == "__main__":
    fix_npz_keys(FILE_PATH)
```
