```python
import shutil
import os

# 1. Дђб»‹nh nghД©a nguб»“n (Drive) vГ  Д‘Г­ch (Local Colab)
# LЖ°u ГЅ: Thay Д‘б»•i Д‘Ж°б»ќng dбє«n nguб»“n chГ­nh xГЎc theo mГЎy cб»§a bбєЎn
drive_cover_dir = "/content/drive/MyDrive/HK1-20252026/Steganography/Data/musdb18-hq"
drive_stego_dir = "/content/drive/MyDrive/HK1-20252026/Steganography/Ablation/7_PhaseCoding"

local_cover_dir = "/content/data/cover"
local_stego_dir = "/content/data/stego"

# 2. HГ m copy nhanh
def copy_data(src, dst):
    if os.path.exists(dst):
        print(f"Deleting old data in {dst}...")
        shutil.rmtree(dst)
    print(f"Copying from {src} to {dst}...")
    shutil.copytree(src, dst)
    print(f"Done! {len(os.listdir(dst))} files copied.")

# 3. Thб»±c hiб»‡n copy
print(">>> ДђANG COPY Dб»® LIб»†U Tб»Є DRIVE XUб»ђNG MГЃY бєўO (SбєЅ mбєҐt vГ i phГєt)...")
try:
    copy_data(drive_cover_dir, local_cover_dir)
    copy_data(drive_stego_dir, local_stego_dir)
    print(">>> COPY HOГЂN Tбє¤T. SбєґN SГЂNG TRAIN!")
except Exception as e:
    print(f"Lб»—i khi copy: {e}")
```
