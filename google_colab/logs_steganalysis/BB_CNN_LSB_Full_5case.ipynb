```python


FILE_PATH = "/content/drive/MyDrive/Git-audio-stego/audio-steganograph/Steganalysis/Data/Impro_LSB_CNN/features_cnn.npz"

def fix_npz_keys(path):
    print(f"--- ДҗANG KIб»ӮM TRA FILE: {path} ---")

    if not os.path.exists(path):
        print(" Lб»—i: KhГҙng tГ¬m thбәҘy file.")
        return

    try:
        # Load file б»ҹ chбәҝ Д‘б»ҷ Д‘б»Қc
        with np.load(path) as data:
            keys = list(data.files)
            print(f" CГЎc key hiб»Үn tбәЎi trong file: {keys}")

            # Nбәҝu Д‘ГЈ Д‘Гәng chuбә©n thГ¬ dб»«ng
            if 'X' in keys and 'y' in keys:
                print(" File Д‘ГЈ Д‘Гәng chuбә©n (cГі X vГ  y). KhГҙng cбә§n sб»ӯa.")
                return

            # Logic Д‘oГЎn vГ  sб»ӯa key
            new_data = {}
            found_X = False
            found_y = False

            for k in keys:
                arr = data[k]
                print(f"   -> Key '{k}': shape={arr.shape}, type={arr.dtype}")

                # ДҗoГЎn X: ThЖ°б»қng lГ  mбәЈng nhiб»Ғu chiб»Ғu (бәўnh: 4 chiб»Ғu, Feature: 2 chiб»Ғu)
                if len(arr.shape) > 1 and not found_X:
                    new_data['X'] = arr
                    found_X = True
                    print(f"      => ДҗГЈ map '{k}' thГ nh 'X'")

                # ДҗoГЎn y: ThЖ°б»қng lГ  mбәЈng 1 chiб»Ғu (Labels)
                elif len(arr.shape) == 1 and not found_y:
                    new_data['y'] = arr
                    found_y = True
                    print(f"      => ДҗГЈ map '{k}' thГ nh 'y'")

            # LЖ°u lбәЎi file mб»ӣi nбәҝu tГ¬m Д‘б»§ X vГ  y
            if found_X and found_y:
                print(" Дҗang ghi Д‘ГЁ file vб»ӣi key chuбә©n 'X' vГ  'y'...")
                np.savez_compressed(path, X=new_data['X'], y=new_data['y'])
                print(" ДҗГғ Sб»¬A THГҖNH CГ”NG! BбәЎn cГі thб»ғ chбәЎy lб»Үnh Train ngay.")
            else:
                print(" KhГҙng thб»ғ tб»ұ Д‘б»ҷng xГЎc Д‘б»Ӣnh X vГ  y. Vui lГІng xГіa file nГ y vГ  tбәЎo lбәЎi cache tб»« Д‘бә§u.")

    except Exception as e:
        print(f" Lб»—i khi Д‘б»Қc file: {e}")

if __name__ == "__main__":
    fix_npz_keys(FILE_PATH)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Load dб»Ҝ liб»Үu
path = "/content/drive/MyDrive/Git-audio-stego/audio-steganograph/Steganalysis/Data/LSB_CNN/features_cnn.npz"
try:
    data = np.load(path)
    print("вң… ДҗГЈ load Д‘Ж°б»Јc file npz.")
    print(f"CГЎc keys trong file: {list(data.keys())}")
except Exception as e:
    print(f"вқҢ Lб»—i khГҙng Д‘б»Қc Д‘Ж°б»Јc file: {e}")
    exit()

# GiбәЈ Д‘б»Ӣnh keys lГ  'x_train'/'y_train' hoбә·c 'X'/'y' tГ№y code cЕ© cб»§a bбәЎn
# Code dЖ°б»ӣi Д‘Гўy tб»ұ Д‘б»ҷng dГІ tГ¬m
keys = list(data.keys())
X_key = [k for k in keys if 'x' in k.lower() or 'feature' in k.lower()][0]
y_key = [k for k in keys if 'y' in k.lower() or 'label' in k.lower()][0]

X = data[X_key]
y = data[y_key]

# 2. Kiб»ғm tra kГӯch thЖ°б»ӣc (Shape)
print("\n--- 1. KIб»ӮM TRA SHAPE ---")
print(f"Shape cб»§a Features (X): {X.shape}")
print(f"Shape cб»§a Labels (y):   {y.shape}")

# 3. Kiб»ғm tra nhГЈn (Labels)
print("\n--- 2. KIб»ӮM TRA NHГғN (LABELS) ---")
unique, counts = np.unique(y, return_counts=True)
print(f"PhГўn bб»‘ nhГЈn: {dict(zip(unique, counts))}")
if len(unique) < 2:
    print("вҡ пёҸ CбәўNH BГҒO Дҗб»Һ: Chб»ү tГ¬m thбәҘy 1 loбәЎi nhГЈn duy nhбәҘt! Data bб»Ӣ lб»—i gГЎn nhГЈn.")

# 4. Kiб»ғm tra giГЎ trб»Ӣ dб»Ҝ liб»Үu (Values)
print("\n--- 3. KIб»ӮM TRA GIГҒ TRб»Ҡ PIXEL/SPECTROGRAM ---")
print(f"Max value: {np.max(X)}")
print(f"Min value: {np.min(X)}")
print(f"Mean value: {np.mean(X)}")
print(f"Sб»‘ lЖ°б»Јng NaN (Not a Number): {np.isnan(X).sum()}")
print(f"Sб»‘ lЖ°б»Јng Inf (VГҙ cб»ұc): {np.isinf(X).sum()}")

# 5. So sГЎnh thб»‘ng kГӘ Cover vs Stego
print("\n--- 4. SO SГҒNH SЖ  Bб»ҳ COVER vs STEGO ---")
# TГЎch ra 2 nhГіm
idx_cover = np.where(y == 0)[0]
idx_stego = np.where(y == 1)[0]

if len(idx_cover) > 0 and len(idx_stego) > 0:
    avg_cover = np.mean(X[idx_cover])
    avg_stego = np.mean(X[idx_stego])
    diff = np.abs(avg_cover - avg_stego)

    print(f"Trung bГ¬nh cб»ҷng pixel Cover: {avg_cover:.6f}")
    print(f"Trung bГ¬nh cб»ҷng pixel Stego: {avg_stego:.6f}")
    print(f"Дҗб»ҷ lб»Үch tuyб»Үt Д‘б»‘i trung bГ¬nh: {diff:.8f}")

    if diff < 1e-5:
        print("вҡ пёҸ CбәўNH BГҒO: Dб»Ҝ liб»Үu Cover vГ  Stego quГЎ giб»‘ng nhau vб»Ғ mбә·t thб»‘ng kГӘ cЖЎ bбәЈn.")
else:
    print("KhГҙng Д‘б»§ dб»Ҝ liб»Үu Д‘б»ғ so sГЎnh.")
```
